```python
import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/TMSL_AIML_ML_PRACTICE/CarPrice_Assignment.csv")
df
```

```python
df.info()
```

```python
x = df.iloc[:,1:-1].values
x
```

```python
y = df.iloc[:,-1].values
y
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
cols_to_encode = [1,2,3,4,5,6,7,8,13,14,16,17]
for i in cols_to_encode:
  x[:,i] = le.fit_transform(x[:,i])
x
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
x_train
```

```python
from sklearn.linear_model import LinearRegression
model =  LinearRegression()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
y_pred = model.predict(x_test)


mae = mean_absolute_error(y_test,y_pred)
print(f"Mean Absolute Error: {mae}")

# r2 = r2_score(y_test,y_pred
mse = mean_squared_error(y_test,y_pred)
print("Mean Squared error : ",mse)
rmse = mse ** (0.5)
print("Root Mean Squared error : ",rmse)

r2 = r2_score(y_test,y_pred)
print("R2 Score : ",r2)
```

```python
import matplotlib.pyplot as plt
plt.scatter(y_test,y_pred,alpha=0.8)
plt.plot(y_test,y_test)
plt.title("Actual vs Predicted")
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.show() # It shows how much predicted values are deviated from original values


```
